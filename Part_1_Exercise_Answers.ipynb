## 1 Use a regular expression to select office files based on file extension

```python
# here is a simple example using a regular expression to find all rows with xls in the Ext colum
df[df.Ext.str.match(r'(xls|doc|xlsx|docx|ppt|pptx)')==True]
```

```python
# Here is another method
#
# Regular expression pattern for office file extensions
pattern = '^(doc|docx|xls|xlsx|ppt|pptx)$'

# Filter rows based on the pattern
df[df['Ext'].str.contains(pattern, case=False, na=False)]
```

## 4 Create a pivot table showing number of files per month based on data last modified

```python
# Set the index to the LastMod column

df_date = df.set_index("LastMod")
```

```python
df_date.pivot_table(index=[df_date.index.year,df_date.index.month],values='Size',aggfunc=['count','sum'])
```

## 5 Create a pivot table listing number of files vs file extension

```python
# run this if you need case sensitivity
df['Ext']=df['Ext'].astype('|S').str.decode("utf-8")
df['Ext']=df['Ext'].str.lower()

pivot_table = df.pivot_table(index='Ext',values='Name',aggfunc=['count'])
print(pivot_table)
```

```python
# You can rename the Name column in the pivot table as follows:

pivot_table.rename(columns={'Name': 'NumberOfFiles'}, inplace=True)
print(pivot_table)
```

```python
# Alternative count like this:

df_counts = df.groupby('Ext').count()['Name']
print(df_counts)
```

```python
# Convert to DataFrame for better formatting
df_counts = df_counts.reset_index()

# Rename the columns for clarity
df_counts.columns = ['FileExtension', 'NumberOfFiles']

# Print the result
print(df_counts)
```

```python
# Let's sort descending on count

sorted_df_counts = df_counts.sort_values(by='NumberOfFiles', ascending=False)
print(sorted_df_counts)
```
